# INDIAN GROCERY STORE PROJECT

# about the data
This dataset contains extensive product listings from a grocery store that targets the Indian market. This dataset includes a rich variety of items, particularly emphasizing organic and culturally specific products. With over 5,700 entries, the dataset provides a deep dive into the pricing, quantity, and category details of products sold, making it a valuable resource for market analysis, consumer behavior studies, and strategic planning in retail

This data comprises of fields like:

                   1. Product name 
                   2. Product category
                   3. Product price in USD
                   4. Product price in INR (indian rupees)
                   5. Quantity
                   
# project objectives 
               1.Market Analysis: Understand pricing strategies and product popularity in Indian grocery sectors.
               2.Consumer Behavior Studies: Analyze preferences for organic and specialty products in the Indian market. 


```python
# importing libraries to load and clean, and visualise the dataset
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')

from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

## exploring and cleaning the data

```python
df = pd.read_csv("C://Users//quays//Downloads//indian_store.csv")
df_copy = df.copy() # making a copy of the original data
```

```python
df.head() # viewing the top 10 rows
```

```python
df.tail(10) # viewing the last 10 rows
```

```python
a, b = df.size, df.shape # checking the size and shape of the data
print(f"the size of the data is {a}\n\nthe shape of the data is {b}")
```

```python
df.duplicated().sum() # checking for duplicated rows 
```

```python
# showing the duplicated rows in the dataset 
duplicates = df[df.duplicated(keep=False)]
print("Duplicate rows:")
# Display the duplicate rows with their indices
duplicates.reset_index(drop=True)
```

```python
df.nunique() # checking for unique values in each column
```

```python
# droping the duplicates 
df.drop_duplicates(inplace = True)
```

```python
# checking for null values 
df.isnull().sum()
```

```python
df.info() #  summary of a dataset structure and information
```

```python
df.isnull().sum()
```

```python
# since Product Price column is the same as product price (INR), i dropped it
df.drop('Product Price', axis = 1, inplace = True)
```

```python
# converting all categorical names to numeric 
df['Product Name'] = encoder.fit_transform(df['Product Name'])
df['Product Category'] = encoder.fit_transform(df['Product Category'])
```

```python
df['Quantity'] = df['Quantity'].str.extract(r'([\d.]+)').astype(float)
```

```python
# new column of product price in usd
df['Price(USD)'] = round((df['Product Price (INR)'] * 0.012 ), 2)
```

# Data visualisation and understanding the data

```python
# heatmap to show columns correlation
plt.figure(figsize = (7, 4))
sns.heatmap(df.corr(), annot = True, cmap = 'coolwarm')
plt.show()
```

```python
plt.figure(figsize = (14, 10))
counts = df_copy['Product Category'].value_counts(ascending=True)
counts.plot(kind = 'barh', color = 'orange')
plt.title('Product popularity bar chart', color = 'black')
plt.ylabel('Product Category', color = 'black')
plt.show()
```

```python
# histogram to see prices in both usd and inr distributions 
plt.figure(figsize = (10, 6))
sns.histplot(df['Product Price (INR)'], kde = True)
plt.title('Product Price (INR) distribution')
plt.show()
```

```python
plt.figure(figsize = (10, 6))
sns.histplot(df['Price(USD)'],binrange=(0, 150), binwidth=10, kde = True)
plt.title('Price(USD) distribution')
plt.show()
```

```python
# histogram to show how the quantity values are distributed
plt.figure(figsize = (10, 6))
sns.histplot(df['Quantity'], kde = True, color = 'green')
plt.title('Quantity distribution')
plt.show()
```
